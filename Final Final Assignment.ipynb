```python
#Brings in today's date.
from datetime import date
today = date.today()
today_date = today.strftime("%m/%d/%y")

#Lines used for display/formatting purposes
solid_line1 = "-"*57
solid_line2 = "="*71

menu = {"Sandwich1": 5.50,
        "Sandwich2" : 6.50,
        "Sandwich3" : 7.00,
        "French Fries" : 3.00,
        "Soft Drink1" : 1.50,
        "Soft Drink2" : 2.00}
#menu_tax = {"Sandwich1": False,
#            "Sandwich2" : False,
#            "Sandwich3" : False,
#            "French Fries" : True,
#            "Fountain Drink1" : True,
#            "Fountain Drink2" : True}
                  
md_name = ['MD','md','Maryland','maryland']
va_name = ['VA','va','Virginia','virginia']
dc_name = ['DC','dc','Washington DC','washington dc','Washington D.C.']        # Lists that allow for more location answers

yes_ans = ['Yes','yes','Y','y','Sure','sure','Yep','yep','You bet','you bet'] 
no_ans = ['No','no','N','n','Nah','nah','No thanks','no thanks']               # Lists that allow for more yes/no answers

customer = input("Welcome to Krusty Burger! Can I get a name for the order? ") # Customer name
acct_no = int(input("Could I get your customer account number? "))             # Customer account number

cust_tax = input("And where are you from? \nMD \nVA \nDC \nOther State ")

# Determines customer state tax rate
def tax_rate(cust_tax) :
    if cust_tax in md_name :
        tax = .06
    elif cust_tax in va_name :
        tax = .0575
    elif cust_tax in dc_name :
        tax = .053
    else :
        tax = .05
    return tax
    
def displayMenu():

    print(f'\n\033[1m{solid_line1}\033[0m')
    
    print("\x1B[4m\x1B[1mMenu:\x1B[0m\x1B[0m") # Underlining and bolding the title. Removed \n
    for k,v in menu.items():
        #print(f"{k[0]+k[1:]} : ${v:.2f}")
        print(f"{k} : ${v:.2f}") #This does the same as code above with fewer characters
        
    print(f'\n\033[1m{solid_line1}\033[0m')
    
def order():
    
    cOrd = []
    total = 0
    
    while True:
        try: 
            #item = str(input("What would you like to order? ")).lower()
            item = str(input("What would you like to order? "))
            menu[item]
            qty = int(input("How many would you like? "))
            price = menu.get(item)
            #print(price)
            cOrd.append((item, price, qty))
            
            tax_rate(cust_tax) # Calling state tax rate function
            
        except:
            print("ERROR: That's not an item on the menu")
        else:    
            print(f'\n\033[1m{solid_line1}\033[0m')
            o2 = str(input("Would you like to order something else? (y/n) ")).lower()

#            if o2 == "y" or o2 == "yes":
            if o2 in yes_ans :
                continue
#            elif o2 == "n" or o2 =="no":
            elif o2 in no_ans :
                
                print(f'\n\033[1m{solid_line1}\033[0m')
                print(f'Customer name:  {customer}')
                print(f'Customer\'s residency: {cust_tax}')
                print(f'Date:  {today_date}')
                print(f'Account No.: {acct_no}')
                print(f'\033[1m{solid_line1}\033[0m')

                print()
                title_bar = f'Item Name \tQuantity \tPrice \t\tCost \t\tTaxable' #created to improve display/format
                tax_able = "Yes"                                                 # Cheating to show a response
                print(f'\033[1m{title_bar}\033[0m')
                print(f'\033[1m{solid_line2}\033[0m')
                
                for item in cOrd:
                    print(f"{item[0].title()}\t{item[2]}\t\t{item[1]:.2f}\t\t{(item[1] * item[2]):.2f} \t\t{tax_able}") #Place this under order
                    total += (item[1] * item[2])
                
                tax_rate(cust_tax) # Calling tax function
                print(f'\n\033[1mSubtotal: ${total:.2f}\033[0m')
                print(f'\033[1mSales tax: ${(total * tax_rate(cust_tax)):.2f}\033[0m')
                print(f'\033[1m{solid_line2}\033[0m')
                print(f'\033[1mTotal: ${(total + (total * tax_rate(cust_tax))):.2f}\033[0m')                
             
                break
            else:
                print("What? Let's try this again...\n")          

while True:
    see = str(input("Would you like to see the menu?(y/n) ")).lower()
    if see in yes_ans :
#    if see == "y" or see == "yes":
        displayMenu()
        break
#    elif see == "n" or see =="no":
    elif see in no_ans :            
        break
    else:
        print("What? Let's try this again...\n")
 
while True:
    o1 = str(input("Would you like to order? (y/n) ")).lower()
#    if o1 == "y" or o1 == "yes":
    if o1 in yes_ans :    
        order()
        break
#    elif o1 == "n" or o1 =="no":
    elif o1 in no_ans:    
        break
    else:
        print("What? Let's try this again...\n")
        
print()
print("Thanks for visiting. Have a krust-astic day!")
```
